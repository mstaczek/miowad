```python
# from MultiLayerPerceptron import NeuralNetwork, Layer
import MultiLayerPerceptron as mlp
net = mlp.NeuralNetwork(weights_random=True)
net.add(mlp.Layer(neurons_count=1, add_bias=False))
net.add(mlp.Layer(neurons_count=1, add_bias=True, activation_fun=mlp.ActivationSigmoid()))
net.add(mlp.Layer(neurons_count=1, add_bias=False, activation_fun=mlp.ActivationLinear()))
# print(net)

net.train([[0],[1]],[[-0.6],[-1]],[[0],[1]],[[-0.6],[-1]],epochs=200,learning_rate=1.6, loss_function=mlp.LossMSE(), regularization=mlp.RegularizationL2(0.003), stop_on_test_set_error_increase=True)
net.predict([[0],[1]]) # around -0.6 and -1
net.plot_training_history()
print(mlp.RegularizationL2(0.003).loss(net.weights))
```
